```python
# Install required packages
!pip install econml dowhy
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LogisticRegression
from econml.dml import LinearDML
from dowhy import CausalModel
import matplotlib.pyplot as plt

# Load and prepare data
url = "https://raw.githubusercontent.com/AMLab-Amsterdam/CEVAE/master/datasets/IHDP/csv/ihdp_npci_1.csv"
columns = ["treatment", "y_factual", "y_cfactual", "mu0", "mu1"] + [f"x{i}" for i in range(1,26)]
df = pd.read_csv(url, header=None)
df.columns = columns

# Split data into features, treatment, and outcome
X = df[[f"x{i}" for i in range(1,26)]]
T = df["treatment"]
Y = df["y_factual"]

# 1. Double Machine Learning Approach (Best Practice)
dml = LinearDML(
    model_y=RandomForestRegressor(n_estimators=100),
    model_t=RandomForestRegressor(n_estimators=100),  # Use Regressor even for binary treatment
    discrete_treatment=True,
    random_state=42
)
dml.fit(Y, T, X=X)
ate_dml = dml.ate(X=X)
print(f"DML ATE: {ate_dml.mean():.2f} (±{ate_dml.std():.2f})")

# 2. Logistic Regression with IPW (Standard Approach)
# Fit propensity score model
lr = LogisticRegression(max_iter=1000)
lr.fit(X, T)
propensity = lr.predict_proba(X)[:, 1]

# Stabilized IPW weights
n_treated = T.sum()
n_control = len(T) - n_treated
weights = (T * n_control/len(T)) / propensity + ((1-T) * n_treated/len(T)) / (1 - propensity)

# Trim extreme weights
q95 = np.quantile(weights, 0.95)
weights = np.where(weights > q95, q95, weights)

# Calculate stabilized IPW ATE
ate_lr = np.average(Y, weights=weights*T) - np.average(Y, weights=weights*(1-T))
print(f"Logistic Regression ATE: {ate_lr:.2f}")

# 3. Causal Model Validation with DoWhy
model = CausalModel(
    data=df,
    treatment="treatment",
    outcome="y_factual",
    common_causes=[f"x{i}" for i in range(1,26)]
)

# Identify and estimate effect
identified_estimand = model.identify_effect()
dml_estimate = model.estimate_effect(
    identified_estimand,
    method_name="backdoor.econml.dml.LinearDML",
    method_params={
        "init_params": {"model_y": RandomForestRegressor(),
                        "model_t": RandomForestRegressor(),
                        "discrete_treatment": True},
        "fit_params": {}
    }
)

# Refutation test
placebo_test = model.refute_estimate(
    identified_estimand,
    dml_estimate,
    method_name="placebo_treatment_refuter",
    placebo_type="permute"
)
print(f"\nPlacebo test result: {placebo_test}")

# Visualization
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.hist(propensity[T==1], bins=20, alpha=0.5, label='Treated')
plt.hist(propensity[T==0], bins=20, alpha=0.5, label='Control')
plt.xlabel('Propensity Score')
plt.title('Propensity Score Distribution')

plt.subplot(1, 2, 2)
plt.scatter(propensity, Y, c=T, alpha=0.5)
plt.xlabel('Propensity Score')
plt.ylabel('Outcome')
plt.title('Treatment Effect Distribution')
plt.show()
```

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt

# ---------------------------
# Load and Prepare the Dataset
# ---------------------------
url = "https://raw.githubusercontent.com/AMLab-Amsterdam/CEVAE/master/datasets/IHDP/csv/ihdp_npci_1.csv"
# Define column names: treatment, factual outcome etc. then 25 covariates
columns = ["treatment", "y_factual", "y_cfactual", "mu0", "mu1"] + [f"x{i}" for i in range(1, 26)]
df = pd.read_csv(url, header=None)
df.columns = columns

# Extract features, treatment indicator, and factual outcome.
# The covariates are x1 to x25.
X = df[[f"x{i}" for i in range(1, 26)]].values.astype(np.float32)
T = df["treatment"].values.astype(np.int32)  # Treatment: assumed 0 or 1
Y = df["y_factual"].values.astype(np.float32)

# ---------------------------
# Prepare Targets and Sample Weights for Multi-Task Learning
# ---------------------------
# We set up two targets:
#   - For the control branch (predicting the outcome if not treated): use Y for T==0 and 0 otherwise.
#   - For the treatment branch (predicting the outcome if treated): use Y for T==1 and 0 otherwise.
y0_target = np.where(T == 0, Y, 0.0)  # outcome for control units; dummy 0 when treated
y1_target = np.where(T == 1, Y, 0.0)  # outcome for treated units; dummy 0 when control

# Similarly, we create sample weights so that each head only contributes loss when its corresponding treatment is observed.
w0 = np.where(T == 0, 1.0, 0.0)  # only weight samples where treatment=0 for control head
w1 = np.where(T == 1, 1.0, 0.0)  # only weight samples where treatment=1 for treatment head

# ---------------------------
# Build a Multi-Task Neural Network (TARNet-like Architecture)
# ---------------------------
# Input layer: 25 covariates
input_layer = keras.Input(shape=(25,), name="covariates")

# Shared representation layers
shared = layers.Dense(64, activation="relu")(input_layer)
shared = layers.Dropout(0.2)(shared)
shared = layers.Dense(32, activation="relu")(shared)

# Control (untreated) head: Predict potential outcome under control
control_branch = layers.Dense(16, activation="relu")(shared)
y0_pred = layers.Dense(1, name="y0")(control_branch)

# Treatment head: Predict potential outcome under treatment
treatment_branch = layers.Dense(16, activation="relu")(shared)
y1_pred = layers.Dense(1, name="y1")(treatment_branch)

# Define the model with two outputs.
model = keras.Model(inputs=input_layer, outputs=[y0_pred, y1_pred])

# Compile the model with mean squared error loss.
model.compile(optimizer=keras.optimizers.Adam(learning_rate=0.001),
              loss="mse")

# ---------------------------
# Train the Model
# ---------------------------
# We train the model so that each head only receives gradient updates for the samples where its outcome is observed.
history = model.fit(X, [y0_target, y1_target],
                    sample_weight=[w0, w1],
                    epochs=100,
                    batch_size=32,
                    validation_split=0.2,
                    verbose=1)

# ---------------------------
# Estimate Treatment Effects
# ---------------------------
# Once training is complete, predict potential outcomes on the entire dataset.
y0_hat, y1_hat = model.predict(X)

# The Individual Treatment Effect (ITE) is estimated as the difference between the treatment and control predictions.
ite = y1_hat.flatten() - y0_hat.flatten()
ate_estimate = np.mean(ite)
print(f"Estimated Average Treatment Effect (ATE): {ate_estimate:.2f}")

# Visualize the distribution of the estimated Individual Treatment Effects (ITE)
plt.figure(figsize=(10, 6))
plt.hist(ite, bins=30, alpha=0.75, color='skyblue', edgecolor='black')
plt.xlabel("Estimated Treatment Effect (y1 - y0)")
plt.ylabel("Frequency")
plt.title("Distribution of Estimated Individual Treatment Effects")
plt.show()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader

# Define a custom dataset class
class IHDPDataset(Dataset):
    def __init__(self, X, T, Y):
        self.X = torch.tensor(X.values, dtype=torch.float32)
        self.T = torch.tensor(T.values, dtype=torch.float32).unsqueeze(-1)
        self.Y = torch.tensor(Y.values, dtype=torch.float32).unsqueeze(-1)

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        return self.X[idx], self.T[idx], self.Y[idx]

# Prepare data
from sklearn.model_selection import train_test_split

X_train, X_val, T_train, T_val, Y_train, Y_val = train_test_split(X, T, Y, test_size=0.2, random_state=42)

# Initialize datasets and data loaders
train_dataset = IHDPDataset(X_train, T_train, Y_train)
val_dataset = IHDPDataset(X_val, T_val, Y_val)

train_data_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_data_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)

# Define a simple neural network model
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(25, 128)  # input layer (25) -> hidden layer (128)
        self.fc2 = nn.Linear(128, 1)  # hidden layer (128) -> output layer (1)

    def forward(self, x):
        x = torch.relu(self.fc1(x))  # activation function for hidden layer
        x = self.fc2(x)
        return x

# Initialize model, optimizer, and loss function
model = Net()
optimizer = optim.Adam(model.parameters(), lr=0.001)
criterion = nn.MSELoss()

# Early stopping class
class EarlyStopping:
    def __init__(self, patience=5, min_delta=0.001):
        self.patience = patience
        self.min_delta = min_delta
        self.counter = 0
        self.best_score = None

    def step(self, val_loss):
        if self.best_score is None:
            self.best_score = val_loss
        elif val_loss > self.best_score + self.min_delta:
            self.counter += 1
            if self.counter >= self.patience:
                return True
        else:
            self.best_score = val_loss
            self.counter = 0
        return False

early_stopping = EarlyStopping(patience=5, min_delta=0.001)

# Train the model
for epoch in range(100):
    # Training loop
    for inputs, _, labels in train_data_loader:
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

    # Calculate validation loss
    model.eval()
    val_loss = 0
    with torch.no_grad():
        for inputs, _, labels in val_data_loader:
            outputs = model(inputs)
            val_loss += criterion(outputs, labels).item()
    val_loss /= len(val_data_loader)

    if early_stopping.step(val_loss):
        print(f"Early stopping at epoch {epoch}")
        break

    print(f"Epoch {epoch+1}, Validation Loss: {val_loss:.4f}")
```

```python
import os
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader

# 1. Load and prepare the IHDP dataset
url = "https://raw.githubusercontent.com/AMLab-Amsterdam/CEVAE/master/datasets/IHDP/csv/ihdp_npci_1.csv"
# The dataset has 5 fixed columns and 25 covariates.
columns = ["treatment", "y_factual", "y_cfactual", "mu0", "mu1"] + [f"x{i}" for i in range(1, 26)]
df = pd.read_csv(url, header=None)
df.columns = columns

# Extract features (x1...x25), treatment and factual outcome.
X = df[[f"x{i}" for i in range(1, 26)]].values.astype(np.float32)
T = df["treatment"].values.astype(np.float32)  # binary: 0 or 1
Y = df["y_factual"].values.astype(np.float32)

# 2. Define a PyTorch Dataset for IHDP
class IHDPDataset(Dataset):
    def __init__(self, X, T, Y):
        self.X = torch.tensor(X)              # shape [n_samples, 25]
        self.T = torch.tensor(T).unsqueeze(1)   # shape [n_samples, 1]
        self.Y = torch.tensor(Y).unsqueeze(1)   # shape [n_samples, 1]
    def __len__(self):
        return len(self.X)
    def __getitem__(self, idx):
        return self.X[idx], self.T[idx], self.Y[idx]

# Create dataset and dataloader
dataset = IHDPDataset(X, T, Y)
dataloader = DataLoader(dataset, batch_size=64, shuffle=True)

# 3. Define the CFR model architecture
class CFRModel(nn.Module):
    def __init__(self, input_dim=25, rep_dim=50, hidden_dim=200):
        super(CFRModel, self).__init__()
        # Representation network: maps input covariates to a latent space
        self.representation = nn.Sequential(
            nn.Linear(input_dim, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, rep_dim),
            nn.ReLU()
        )
        # Two separate heads for each treatment group:
        self.head0 = nn.Sequential(
            nn.Linear(rep_dim, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, 1)
        )
        self.head1 = nn.Sequential(
            nn.Linear(rep_dim, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, 1)
        )
    def forward(self, x, t):
        # x: covariates, t: treatment indicator (0 or 1)
        z = self.representation(x)  # latent representation
        y0_pred = self.head0(z)     # outcome if t==0
        y1_pred = self.head1(z)     # outcome if t==1
        # Select output based on treatment: pred = t*y1_pred + (1-t)*y0_pred.
        pred = t * y1_pred + (1 - t) * y0_pred
        return pred, z, y0_pred, y1_pred

# 4. Define an MMD Loss function (our IPM penalty) using pairwise squared distances.
def mmd_loss(x, y, kernel_bandwidth=1.0):
    # x: tensor of shape [n_x, d]
    # y: tensor of shape [n_y, d]
    # Compute squared norms for each row.
    x_norm = torch.sum(x ** 2, dim=1, keepdim=True)  # shape [n_x, 1]
    y_norm = torch.sum(y ** 2, dim=1, keepdim=True)  # shape [n_y, 1]

    # dxx: pairwise squared distances for x.
    dxx = x_norm + x_norm.t() - 2 * torch.mm(x, x.t())
    # dyy: pairwise squared distances for y.
    dyy = y_norm + y_norm.t() - 2 * torch.mm(y, y.t())
    # dxy: pairwise squared distances between x and y.
    dxy = x_norm + y_norm.t() - 2 * torch.mm(x, y.t())

    Kxx = torch.exp(-dxx / (2 * kernel_bandwidth))
    Kyy = torch.exp(-dyy / (2 * kernel_bandwidth))
    Kxy = torch.exp(-dxy / (2 * kernel_bandwidth))

    mmd = Kxx.mean() + Kyy.mean() - 2 * Kxy.mean()
    return mmd

# 5. Instantiate the model, optimizer, and loss function.
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = CFRModel(input_dim=25, rep_dim=50, hidden_dim=200).to(device)
optimizer = optim.Adam(model.parameters(), lr=1e-3)
mse_loss = nn.MSELoss(reduction="none")  # will apply per-sample weights

# Compute global treatment proportion u and pre-compute sample weights for full dataset.
u = T.mean()  # global fraction treated (a scalar)
global_weights = (T / (2 * u)) + ((1 - T) / (2 * (1 - u)))
global_weights = torch.tensor(global_weights, dtype=torch.float32).unsqueeze(1).to(device)

# 6. Training loop
num_epochs = 100
alpha = 1.0  # Regularization coefficient for IPM penalty

for epoch in range(num_epochs):
    model.train()
    epoch_loss = 0.0
    for batch in dataloader:
        x_batch, t_batch, y_batch = [b.to(device) for b in batch]

        # Forward pass.
        y_pred, z, y0_pred, y1_pred = model(x_batch, t_batch)

        # Factual loss: weighted mean squared error between prediction and observed outcome.
        loss_per_sample = mse_loss(y_pred, y_batch)
        batch_weights = t_batch / (2 * u) + (1 - t_batch) / (2 * (1 - u))
        factual_loss = torch.mean(batch_weights * loss_per_sample)

        # Compute IPM penalty using MMD on the latent representations.
        z_treated = z[(t_batch == 1).squeeze()]
        z_control = z[(t_batch == 0).squeeze()]
        ipm_pen = 0.0
        if z_treated.size(0) > 1 and z_control.size(0) > 1:
            ipm_pen = mmd_loss(z_treated, z_control, kernel_bandwidth=1.0)

        total_loss = factual_loss + alpha * ipm_pen

        optimizer.zero_grad()
        total_loss.backward()
        optimizer.step()

        epoch_loss += total_loss.item() * x_batch.size(0)

    avg_loss = epoch_loss / len(dataset)
    print(f"Epoch {epoch+1:03d}/{num_epochs} - Loss: {avg_loss:.4f}")

# 7. Evaluation: Compute predicted ITE and ATE over the entire dataset.
model.eval()
with torch.no_grad():
    X_tensor = torch.tensor(X, dtype=torch.float32).to(device)
    t0 = torch.zeros((X_tensor.size(0), 1)).to(device)
    t1 = torch.ones((X_tensor.size(0), 1)).to(device)

    # Get predictions for both treatment values.
    _, _, y0_pred_all, _ = model(X_tensor, t0)
    _, _, _, y1_pred_all = model(X_tensor, t1)

    # Individual treatment effect (ITE): difference between treated and control predictions.
    ite_pred = y1_pred_all - y0_pred_all
    # Average treatment effect (ATE): average of the ITE over the sample.
    ate_pred = ite_pred.mean().item()
    print("\nEstimated Average Treatment Effect (ATE): {:.4f}".format(ate_pred))
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from scipy.stats import norm

# Load IHDP dataset
url = "https://raw.githubusercontent.com/AMLab-Amsterdam/CEVAE/master/datasets/IHDP/csv/ihdp_npci_1.csv"
columns = ["treatment", "y_factual", "y_cfactual", "mu0", "mu1"] + [f"x{i}" for i in range(1,26)]
df = pd.read_csv(url, header=None)
df.columns = columns

# Extract variables
X = df[[f"x{i}" for i in range(1,26)]]  # Covariates
T = df["treatment"]                     # Treatment variable (binary: 0 or 1)
Y = df["y_factual"]                      # Observed outcome

# Standardize covariates for logistic regression
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

### 📌 STEP 1: Estimate Propensity Scores using Logistic Regression
log_reg = LogisticRegression(max_iter=1000)
log_reg.fit(X_scaled, T)  # Fit logistic regression model
propensity_scores = log_reg.predict_proba(X_scaled)[:, 1]  # P(T=1|X)

# 📌 STEP 2: Clip propensity scores to avoid extreme values
propensity_scores = np.clip(propensity_scores, 0.1, 0.9)

# 📌 STEP 3: Compute Inverse Probability Weights (IPW)
weights = (T / propensity_scores) + ((1 - T) / (1 - propensity_scores))

# 📌 STEP 4: Estimate ATE using IPW
ate_ipw = (Y * T / propensity_scores).mean() - (Y * (1 - T) / (1 - propensity_scores)).mean()
print(f"IPW Estimated ATE: {ate_ipw:.2f}")

# 📌 STEP 5: Check Covariate Balance using Standardized Mean Differences (SMD)
def standardized_mean_difference(X, T, weights):
    smd_values = []
    for i in range(X.shape[1]):
        X_treated = X[T == 1, i]
        X_control = X[T == 0, i]

        mean_treated = np.average(X_treated, weights=weights[T == 1])
        mean_control = np.average(X_control, weights=weights[T == 0])

        var_treated = np.average((X_treated - mean_treated) ** 2, weights=weights[T == 1])
        var_control = np.average((X_control - mean_control) ** 2, weights=weights[T == 0])

        smd = abs(mean_treated - mean_control) / np.sqrt((var_treated + var_control) / 2)
        smd_values.append(smd)

    return smd_values

smd_values = standardized_mean_difference(X_scaled, T, weights)
max_smd = max(smd_values)
print(f"Max Standardized Mean Difference after IPW: {max_smd:.4f}")

# 📌 Interpretation: If max SMD < 0.1, covariates are balanced
if max_smd < 0.1:
    print("Covariates are well-balanced after IPW.")
else:
    print("Covariates may still be imbalanced.")
```

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_squared_error

# Load IHDP dataset
url = "https://raw.githubusercontent.com/AMLab-Amsterdam/CEVAE/master/datasets/IHDP/csv/ihdp_npci_1.csv"
columns = ['T'] + [f'X{i}' for i in range(1, 26)] + ['y_factual', 'y_cfactual', 'mu0', 'mu1']
data = pd.read_csv(url, header=None, names=columns)

# Extract relevant columns
T = data['T'].values  # Treatment assignment (Binary: 0 or 1)
Y = data['y_factual'].values  # Observed outcome
X = data.drop(columns=['T', 'y_factual', 'mu0', 'mu1']).values  # Covariates
ATE_true = np.mean(data['mu1'] - data['mu0'])  # True ATE from IHDP

# Step 1: Train first-stage models (Outcome & Propensity Score)
rf_Y = RandomForestRegressor(n_estimators=100, random_state=42).fit(X, Y)
rf_T = LogisticRegression().fit(X, T)  # Logistic regression for treatment model

# Step 2: Compute predicted values
Y_hat = rf_Y.predict(X)  # Predicted outcomes
T_hat = np.clip(rf_T.predict_proba(X)[:, 1], 0.01, 0.99)  # Clip probabilities

# Step 3: Compute residualized outcomes and treatments
Y_res = Y - Y_hat  # Residualized outcome
T_res = T - T_hat  # Residualized treatment

# Step 4: Estimate ATE using residualized outcomes
ATE_hat = np.mean(Y_res * T_res) / np.mean(T_res ** 2)

# Step 5: Compute evaluation metrics
MSE_Y = mean_squared_error(Y, Y_hat)  # MSE of outcome model
MSE_ATE = (ATE_hat - ATE_true) ** 2  # MSE of ATE
Bias_ATE = abs(ATE_hat - ATE_true)  # Bias

# Confidence Interval for ATE
n = len(Y)
ATE_std_error = np.std(Y_res * T_res) / np.sqrt(n)
CI_lower = ATE_hat - 1.96 * ATE_std_error
CI_upper = ATE_hat + 1.96 * ATE_std_error

# Step 6: Print Results
print(f"MSE of Outcome Model: {MSE_Y:.4f}")
print(f"Estimated ATE: {ATE_hat:.4f}")
print(f"True ATE: {ATE_true:.4f}")
print(f"MSE of ATE Estimation: {MSE_ATE:.4f}")
print(f"Bias of ATE: {Bias_ATE:.4f}")
print(f"95% Confidence Interval for ATE: ({CI_lower:.4f}, {CI_upper:.4f})")
```

```python
import numpy as np
import pandas as pd
import requests
from io import StringIO
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, mean_squared_error

# Load IHDP dataset
url = "https://raw.githubusercontent.com/AMLab-Amsterdam/CEVAE/master/datasets/IHDP/csv/ihdp_npci_1.csv"
response = requests.get(url)
data = pd.read_csv(StringIO(response.text), header=None)

# Extract relevant variables
y = data.iloc[:, 0]  # Treatment indicator (0: control, 1: treated)
x = data.iloc[:, 1:-1]  # Covariates
true_ate = 4.01606689611833

# Fit a propensity score model
ps_model = LogisticRegression()
ps_model.fit(x, y)
propensity_scores = ps_model.predict_proba(x)[:, 1]

# Compute Inverse Probability Weighting (IPW) estimates
weights = y / propensity_scores + (1 - y) / (1 - propensity_scores)
ipw_ate = np.mean(weights * y) - np.mean(weights * (1 - y))

# Compute Bias and MSE
bias = ipw_ate - true_ate
relative_bias = (bias / true_ate) * 100
mse = mean_squared_error([true_ate], [ipw_ate])

# Compute classification metrics
threshold = 0.25
y_pred = (propensity_scores >= threshold).astype(int)
accuracy = accuracy_score(y, y_pred)
precision = precision_score(y, y_pred)
recall = recall_score(y, y_pred)
f1 = f1_score(y, y_pred)
roc_auc = roc_auc_score(y, propensity_scores)
conf_matrix = confusion_matrix(y, y_pred)

# Print results
print(f"Estimated ATE (IPW): {ipw_ate:.4f}")
print(f"True ATE: {true_ate:.4f}")
print(f"Bias: {bias:.4f}")
print(f"Relative Bias: {relative_bias:.2f}%")
print(f"MSE: {mse:.4f}")
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
print(f"ROC AUC: {roc_auc:.4f}")
print("Confusion Matrix:")
print(conf_matrix)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, auc

from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from econml.dml import LinearDML

# Load IHDP dataset
url = "https://raw.githubusercontent.com/AMLab-Amsterdam/CEVAE/master/datasets/IHDP/csv/ihdp_npci_1.csv"
columns = ["treatment", "y_factual", "y_cfactual", "mu0", "mu1"] + [f"x{i}" for i in range(1, 26)]
df = pd.read_csv(url, header=None)
df.columns = columns

# Prepare data: covariates (X), treatment indicator (T), and observed outcome (Y)
X = df[[f"x{i}" for i in range(1, 26)]].values.astype(np.float32)
T = df["treatment"].values.astype(np.float32)
Y = df["y_factual"].values.astype(np.float32)

# True ATE from IHDP (benchmark)
true_ATE = np.mean(df["mu1"] - df["mu0"])
print(f"True ATE: {true_ATE:.4f}")

# --- Double Machine Learning (DML) Approach ---
dml = LinearDML(
    model_y=RandomForestRegressor(n_estimators=100, random_state=42),
    model_t=RandomForestClassifier(n_estimators=100, random_state=42),
    discrete_treatment=True,
    random_state=42
)
dml.fit(Y, T, X=X)
ate_dml = dml.ate(X=X)
print(f"DML ATE: {ate_dml.mean():.2f} (±{ate_dml.std():.2f})")

# --- Propensity Score Model Evaluation for DML ---
# Note: dml.model_t may not be directly accessible, so we fit a separate logistic regression.
lr = LogisticRegression(max_iter=1000)
lr.fit(X, T)
propensity_scores = lr.predict_proba(X)[:, 1]

# Compute ROC curve and AUC using the separately fitted logistic regression model
fpr, tpr, thresholds = roc_curve(T, propensity_scores)
roc_auc = auc(fpr, tpr)
print(f"AUC (DML Treatment Model via separate LR): {roc_auc:.4f}")

# Plot ROC curve with optimal threshold (e.g., 0.25) marked
optimal_threshold = 0.25
optimal_idx = np.argmin(np.abs(thresholds - optimal_threshold))

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', lw=2, label=f"ROC Curve (AUC = {roc_auc:.4f})")
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.scatter(fpr[optimal_idx], tpr[optimal_idx], color='red', s=80,
            label=f"Optimal Threshold ({optimal_threshold})")
plt.xlabel("False Positive Rate (FPR)")
plt.ylabel("True Positive Rate (TPR)")
plt.title("ROC Curve for DML Treatment Model")
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```
